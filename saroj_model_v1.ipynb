```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv('dataset.csv')
dataset.head()
```

```python
dataset.describe()
```

Converting B to 0 and M to 1

```python
dataset.drop('id',axis=1,inplace=True)
dataset['diagnosis'].replace('B', 0,inplace=True)
dataset['diagnosis'].replace('M', 1,inplace=True)
dataset.head()
```

Plotting the heatmap

```python
sns.heatmap(dataset[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','points_mean','symmetry_mean','dimension_mean']].corr())
```

Scaling data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(dataset.drop(('diagnosis'),axis=1))
scaled=scaler.transform(dataset.drop(('diagnosis'),axis=1))
df=pd.DataFrame(scaled,columns=dataset.columns[:-1])
```

#### Training model

```python
X=df
y=dataset['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=20)
```

####  Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
pknn=knn.predict(X_test)
```

Finding accuracy

```python
from sklearn.metrics import classification_report,confusion_matrix
```

Accuracy of Knn

```python
print(classification_report(y_test,pknn))
```

The performance decreases on addition of other parameters.
